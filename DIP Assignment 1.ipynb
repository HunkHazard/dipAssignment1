```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("coins.jpg",)

# Task 1: Resize the image to 256x256
image_resized = cv2.resize(image, (256, 256))
# channels are reversed in cv2. cv2 uses BGR instead of RGB and matplotlib uses RGB
plt.imshow(cv2.cvtColor(image_resized, cv2.COLOR_BGR2RGB))
```

```python
# Task 2: Convert the image to rgb format
image_rgb = cv2.cvtColor(image_resized,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
# Task 3: Convert the image to grayscale
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray,cmap="gray")
```

```python
# Task 4: Convert the image to binary
threshold_value = 200

_, binary_image = cv2.threshold(image_gray, threshold_value, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap="gray")
```

```python
# Task 5: Find the number of coins in the image

# step 1: convert the image to grayscale
plt.imshow(image_gray,cmap="gray")
```

```python
# step 2: blur the image
blurred_image = cv2.GaussianBlur(image_gray, (5, 5), 0)
plt.imshow(blurred_image,cmap="gray")
```

```python
# step 3: find edges

canny_image = cv2.Canny(blurred_image, 30, 150)
plt.imshow(canny_image,cmap="gray")
```

```python
# step 4: dilate the image (to make the edges thicker)
dilated_image = cv2.dilate(canny_image, (5, 5), iterations=1)
plt.imshow(dilated_image,cmap="gray")
```

```python
# step 5: find contours
contours, hierarchy = cv2.findContours(dilated_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
image_copy = image_rgb.copy()
cv2.drawContours(image_copy, contours, -1, (0, 255, 0), 2)
plt.imshow(image_copy)
```

```python
# step 6: print the number of coins
print("Number of coins in the image: {}".format(len(contours)))
```
